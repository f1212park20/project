```python
# 파이썬의 multiprocessing 라이브러리를 사용하면 여러 개의 프로세스를 동시에 실행하여 작업 속도를 크게 향상시킬 수 있습니다.
# 병렬 처리: 병렬 처리는 여러 개의 작업을 동시에 실행하여 프로그램의 성능을 향상시키는 기술
from concurrent.futures import ThreadPoolExecutor
import pandas as pd

# 인덱스 구성종목 조회
# 코스피 200
pdf = stock.get_index_portfolio_deposit_file("1028")
# print(len(pdf), pdf)

# 다음은 20210101-20210528 동안의 "코스피 200" 인덱스의 OHLCV를 조회
# df = stock.get_index_ohlcv("20210101", "20210528", "1028")
# print(df.head(2))

# 다음은 20210101-20210528의 "코스피 200"의 월별 OHLCV를 출력합니다.
# df = stock.get_index_ohlcv("20210101", "20210528", "1028", "m")
# print(df.head(2))

# 변동률
# 등락률:특정 기간 동안의 값의 변화를 백분율로 나타낸 것으로, 변동률과 가장 흔히 혼용되는 단어입니다. 주가, 환율 등에서 많이 사용됩니다.
# 1.for문을 통해 조회
# for i in pdf:
#   df = stock.get_market_ohlcv("20210201", "20210201", i)
#   print(df.head())

def get_ohlcv(ticker):
  df = stock.get_market_ohlcv("20210201", "20210201", ticker)
  df['종목명'] = stock.get_market_ticker_name(ticker) # 종목명 추가
  return df.head()

with ThreadPoolExecutor() as executor:
  results = list(executor.map(get_ohlcv, pdf))

for result in results:
  print(result)

df = pd.concat(results, keys=pdf)
print(df)
```
